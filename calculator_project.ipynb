```python
import math

def calculator():
    print("Standard Calculator")
    try:
        while True:
            op = input("Enter operation (+, -, *, /, sin, cos, square, or exit): ").lower()

            if op == "exit":
                print("Exiting the calculator...")
                break

            basic_ops = ["+", "-", "*", "/"]
            advanced_ops = ["sin", "cos", "square"]

            if op in basic_ops:
                numbers = []
                while True:
                    number = input("Enter number or '=' to calculate: ")
                    if number == "=":
                        if len(numbers) < 2:
                            print("Enter at least 2 numbers.")
                            continue
                        break
                    try:
                        num = float(number)
                        numbers.append(num)
                    except ValueError:
                        print("Enter numbers only!")
                        continue

                result = numbers[0]
                for number in numbers[1:]:
                    if op == "+":
                        result += number
                    elif op == "-":
                        result -= number
                    elif op == "*":
                        result *= number
                    elif op == "/":
                        if number == 0:
                            print("Error: Division by zero.")
                            break
                        result /= number

                expression = f" {op} ".join(map(str, numbers))
                print(f"{expression} = {result}")

            elif op in advanced_ops:
                number = input("Enter a number: ")
                try:
                    num = float(number)
                    if op == "sin":
                        result = math.sin(math.radians(num))
                        print(f"sin({num}) = {result}")
                    elif op == "cos":
                        result = math.cos(math.radians(num))
                        print(f"cos({num}) = {result}")
                    elif op == "square":
                        result = num ** 2
                        print(f"{num}Â² = {result}")
                except ValueError:
                    print("Enter a valid number!")

            else:
                print("Please enter a valid operator.")

    except Exception as e:
        print(f"Invalid operation: {e}")

    print("Developed by: Hayat")

# Run the calculator
calculator()
```
